```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/data_science_job.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.isnull().mean()*100
```

```python
df.shape
```

```python
good_cols = [i for i in df.columns if df[i].isnull().mean()<0.05 and df[i].isnull().mean()>0]
```

```python
df[good_cols].sample(6)
```

```python
df['education_level'].value_counts()
```

```python
df['education_level'].isnull().sum()
```

```python
len(df[good_cols].dropna()) / len(df)
```

```python
newdf = df[good_cols].dropna()
```

```python
newdf.shape, df.shape
```

```python
sns.distplot(newdf['experience'])
```

```python
sns.distplot(df['experience'])
```

```python
plt.subplot(1,2,1)
sns.distplot(newdf['city_development_index'])
plt.subplot(1,2,2)
sns.distplot(df['city_development_index'])
```

```python
plt.subplot(1,2,1)
sns.distplot(newdf['training_hours'])
plt.subplot(1,2,2)
sns.distplot(df['training_hours'])
```

```python
sns.kdeplot(df['city_development_index'])
sns.kdeplot(newdf['city_development_index'])
```

```python
sns.kdeplot(df['training_hours'],color='r',alpha=0.7)
sns.kdeplot(newdf['training_hours'],color='b')
```

# For Categorical Data

```python
temp = pd.concat([
            # percentage of observations per category, original data
            df['enrolled_university'].value_counts() / len(df),

            # percentage of observations per category, cca data
            newdf['enrolled_university'].value_counts() / len(newdf)
        ],axis=1)

# add column names
temp.columns = ['original', 'cca']

temp
```

```python
temp = pd.concat([
            # percentage of observations per category, original data
            df['education_level'].value_counts() / len(df),

            # percentage of observations per category, cca data
            newdf['education_level'].value_counts() / len(newdf)
        ],axis=1)

# add column names
temp.columns = ['original', 'cca']

temp
```
