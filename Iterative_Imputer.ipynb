```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df = np.round(pd.read_csv('50_Startups.csv')[['R&D Spend','Administration','Marketing Spend','Profit']]/10000)
np.random.seed(9)
df = df.sample(5)
df
```

```python
df = df.iloc[:,0:-1]
df
```

```python
df.iloc[1,0] = np.nan
df.iloc[3,1] = np.nan
df.iloc[-1,-1] = np.nan
```

```python
df
```

```python

df0 = pd.DataFrame()

df0['R&D Spend'] = df['R&D Spend'].fillna(df['R&D Spend'].mean())
df0['Administration'] = df['Administration'].fillna(df['Administration'].mean())
df0['Marketing Spend'] = df['Marketing Spend'].fillna(df['Marketing Spend'].mean())
```

```python
df0
```

```python
df1 = df0.copy()

df1.iloc[1,0] = np.nan

df1
```

```python

y = df1.iloc[[0,2,3,4],0]
y
```

```python
lr = LinearRegression()
lr.fit(X,y)
lr.predict(df1.iloc[1,1:].values.reshape(1,2))
```

```python
df2 = df1.copy()

df2.iloc[1,0] = np.nan

df2
```

```python
X = df2.iloc[[0,2,3,4],1:3]
y = df2.iloc[[0,2,3,4],0]

lr = LinearRegression()
lr.fit(X,y)
lr.predict(df2.iloc[1,1:].values.reshape(1,2))
```

```python
df2.iloc[3,1] = 11.22
```

```python
df2.iloc[4,-1] = np.nan

# Filter out rows that have NaN values in 'R&D Spend', 'Administration', or 'Marketing Spend'
# These are the columns used for training to predict 'Marketing Spend'
rows_for_training = df2.iloc[[0,2,3]] # Rows 0, 2, 3 have no NaNs in these columns

X_train = rows_for_training.iloc[:,0:2] # Features: 'R&D Spend', 'Administration'
y_train = rows_for_training.iloc[:,-1] # Target: 'Marketing Spend'

lr = LinearRegression()
lr.fit(X_train, y_train)

# Predict the missing Marketing Spend for row 4 using its R&D Spend and Administration values
predicted_value = lr.predict(df2.iloc[4,0:2].values.reshape(1,2))[0]
df2.iloc[4,-1] = predicted_value

df2
```
