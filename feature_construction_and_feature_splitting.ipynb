```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('/content/train (1).csv')
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder

# Select only numerical features for X
X = df[['PassengerId', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Survived']]

# Encode the categorical target variable y (Embarked) into numerical labels
le = LabelEncoder()
y = le.fit_transform(df['Embarked'])
```

```python
X
```

```python
y
```

```python
# Scale the features in X
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform cross-validation with scaled data and adjusted n_splits
# n_splits is reduced to 5 to avoid issues with very small classes
np.mean(cross_val_score(LogisticRegression(max_iter=1000), X_scaled, y, scoring='accuracy', cv=5))
```

Applying Feature Construction

```python
X['Family_size'] = X['SibSp'] + X['Parch'] + 1
```

```python
X
```

```python
def myfunc(num):
    if num == 1:
        #alone
        return 0
    elif num >1 and num <=4:
        # small family
        return 1
    else:
        # large family
        return 2
```

```python
X['Family_type'] = X['Family_size'].apply(myfunc)
```

```python
X
```

```python
X.drop(columns=['SibSp','Parch','Family_size'],inplace=True)
```

```python
X
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))
```
